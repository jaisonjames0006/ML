```python
catpath=os.path.join('/content/drive/MyDrive/Data','Cat')
for img in os.listdir(catpath):
  print(img)
```

```python
flat_data_arr=[]
target_arr=[]
Categories=['Cat','Dog'] #Cat=0, Dog=1
```

```python
datadir='/content/drive/MyDrive/Data'

for i in Categories:
  print('loading... category',i)
  path=os.path.join(datadir,i)   #i=Cat
  for img in os.listdir(path):   #/content/drive/MyDrive/Data/Cat
    img_array=imread(os.path.join(path,img)) # to read image in pixel array
    img_resized=resize(img_array,(150,150,3))
    flat_data_arr.append(img_resized.flatten()) # flatten is used to convert array to 1D
    target_arr.append(Categories.index(i))
  print('loaded',i)
```

```python
df=pd.DataFrame(flat_data_arr)
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.fit_transform(x_test)
```

```python
from sklearn.svm import SVC
clf=SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
print('classification_report\n',classification_report(y_test,y_pred))
print('confusion matrix\n',confusion_matrix(y_test,y_pred))
print('Display\n',ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
